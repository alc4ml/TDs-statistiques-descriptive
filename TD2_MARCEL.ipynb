On peut parfois observer que certains clusters se rapprochent de certaines phases (ex. croisiÃ¨re ou descente),  
mais la correspondance nâest pas exacte.


```python
import numpy as np

def cluster_purity(labels, targets):
    """
    Calcule la puretÃ© du clustering selon la formule :
        purity = (1/N) * Î£_k max_j |Q_k â© Y_j|
    labels  : Ã©tiquettes de clusters (KMeans)
    targets : Ã©tiquettes vraies (ex: phases de vol)
    """
    # Conversion en array numpy pour Ã©viter les erreurs
    labels = np.array(labels)
    targets = np.array(targets)

    # Nombre total de points
    N = len(labels)

    # PuretÃ© cumulÃ©e
    total = 0

    # Pour chaque cluster k, on trouve la classe la plus frÃ©quente
    for cluster in np.unique(labels):
        # points appartenant Ã  ce cluster
        mask = labels == cluster
        # classes correspondantes dans la cible
        target_in_cluster = targets[mask]
        # frÃ©quence de chaque classe
        most_common = np.bincount(target_in_cluster).max()
        total += most_common

    return total / N
```

```python
from sklearn.preprocessing import LabelEncoder

# Encodage des phases de vol
encoder = LabelEncoder()
targets_encoded = encoder.fit_transform(phase_labels)
```

```python
# Liste des valeurs de K Ã  tester
K_values = [1, 2, 3, 5, 10, 20, 30, 50]
purity_scores = []

for k in K_values:
    # K-means pour chaque K
    kmeans = KMeans(n_clusters=k, random_state=0)
    kmeans.fit(X_vol_norm)
    clusters = kmeans.labels_

    # Calcul de la puretÃ©
    pur = cluster_purity(clusters, targets_encoded)
    purity_scores.append(pur)
```

```python
plt.figure(figsize=(8,6))
plt.plot(K_values, np.array(purity_scores)*100, marker='o')
plt.title("Ãvolution de la puretÃ© en fonction du nombre de clusters K")
plt.xlabel("Nombre de clusters (K)")
plt.ylabel("PuretÃ© (%)")
plt.grid(True)
plt.show()
```

**InterprÃ©tation :**

La puretÃ© augmente avec le nombre de clusters K, car plus K est grand, plus les groupes sont petits et homogÃ¨nes.  
Cependant, Ã  partir de **K â 10 Ã  20**, la puretÃ© se stabilise,  
ce qui signifie que crÃ©er davantage de clusters nâapporte plus dâamÃ©lioration significative.

La "meilleure" valeur de K se situe donc autour de **K = 10 Ã  20**


```python
from scipy.optimize import linear_sum_assignment
```

```python
def cluster_acc(labels, targets):
    assert labels.size == targets.size
    D = max(labels.max(), targets.max()) + 1
    w = np.zeros((D, D))
    for i in range(labels.size):
        w[labels[i], targets[i]] += 1
    row_ind, col_ind = linear_sum_assignment(w.max() - w)
    return w[row_ind, col_ind].sum() * 1.0 / labels.size

acc = cluster_acc(labels, vol['FLIGHT_MOD'])
print('Accuracy : {:f} %'.format(acc*100))
```
